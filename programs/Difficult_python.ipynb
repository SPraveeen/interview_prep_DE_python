```python
def replace_cube(arr,N):
    answer=[]

    for i in range(len(arr)):
        if ((i+1)%N==0):
            arr[i]=arr[i]**3
        answer.append(arr[i])
    return answer
```

```python
a=replace_cube([1,2,3,4,2,5],3)
print(a)
```

```python
def flatten(arr):
    answer=[]
    while arr:
        element=arr.pop()
        if type(element)==list:
            arr.extend(element)
        else:
            answer.append(element)
        answer.sort()
    return answer

# pop,extend,append
```

```python
flatten([1,2,3,[4,5,6],7,8])
```

```python
def new_menu(item_list):
    #This is the Menu list we want to update
    Menu = [['Sweet_Corn_Soup', 300.0], 
    ['Cream_of_Tomato_Soup', 100.0], 
    ['Bacon_and_Cheese', 150.0], 
    ['Honey_Mustard', 230.0], 
    ['Hot_Coffee', 50.0], 
    ['Cold_Coffee', 50.0], 
    ['Egg_Sandwiches', 130.0], 
    ['Tacos', 400.0]]

    for i in item_list:
        for j in Menu:
            if i in j:
                j[1]+=0.1*j[1]
    return Menu
```

```python
print(new_menu(['Hot_Coffee','Cold_Coffee','Tacos']))
```

```python
import math
def lcm(arr):    
    result=1
    for num in arr:
        result=abs(result*num)//math.gcd(result,num)
    return result
```

```python
print(lcm([13,6,17,18,19,20,37]))
```
