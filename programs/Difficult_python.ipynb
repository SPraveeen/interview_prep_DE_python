```python
def replace_cube(arr,N):
    answer=[]

    for i in range(len(arr)):
        if ((i+1)%N==0):
            arr[i]=arr[i]**3
        answer.append(arr[i])
    return answer
```

```python
a=replace_cube([1,2,3,4,2,5],3)
print(a)
```

```python
def flatten(arr):
    answer=[]
    while arr:
        element=arr.pop()
        if type(element)==list:
            arr.extend(element)
        else:
            answer.append(element)
        answer.sort()
    return answer

# pop,extend,append
```

```python
flatten([1,2,3,[4,5,6],7,8])
```

```python
def new_menu(item_list):
    #This is the Menu list we want to update
    Menu = [['Sweet_Corn_Soup', 300.0], 
    ['Cream_of_Tomato_Soup', 100.0], 
    ['Bacon_and_Cheese', 150.0], 
    ['Honey_Mustard', 230.0], 
    ['Hot_Coffee', 50.0], 
    ['Cold_Coffee', 50.0], 
    ['Egg_Sandwiches', 130.0], 
    ['Tacos', 400.0]]

    for i in item_list:
        for j in Menu:
            if i in j:
                j[1]+=0.1*j[1]
    return Menu
```

```python
print(new_menu(['Hot_Coffee','Cold_Coffee','Tacos']))
```

```python
import math
def lcm(arr):    
    result=1
    for num in arr:
        result=abs(result*num)//math.gcd(result,num)
    return result
```

```python
print(lcm([13,6,17,18,19,20,37]))
```

```python
def compute_dist(list,pnt):
    x0,y0=pnt
    ans=[]
    for point in list:
        x,y=point
        dist=(x-x0)**2+(y-y0)**2
        ans.append((dist,point))

    for i in range(len(ans)):
        for j in range(i+1,len(ans)):
            if ans[i][0]>ans[j][0]:
                ans[i],ans[j]=ans[j],ans[i]
    
    result=[]
    for i in range(min(5,len(ans))):
        result.append(ans[i][1])
    return result
```

```python
lis = [(7,8),(1,1),(2,2),(3,3),(4,4),(6,7)]
pnt = (0,0)
print(compute_dist(lis,pnt))
```

```python
def count_unique_words(inp):
    ans=0
    word=inp.split()
    for words in word:
        if len(words)>3 and len(words)==len(set(words)):
            ans+=1
    return (ans)
```

```python
print(count_unique_words('Hii Hello Helo Hey'))
print(count_unique_words('the fruits of the tree were basically oranges and apples'))
```

```python
from collections import Counter

def most_frequent(arr):

    freq=Counter(arr)

    most_frequent_element=None
    max_frequency=0

    for element,count in freq.items():
        if count>max_frequency:
            most_frequent_element=element
            max_frequency=count

    return most_frequent_element
```

```python
print(most_frequent(['Hii','Hello','Hii']))
```
