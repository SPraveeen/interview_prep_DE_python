- add numbers

- this time completestity was o(n) as input incresases steps increas

```python
arr=[1,2,3,4,5,6]

def get_sum(arr):
    ans=0
    iteration=0
    for num in arr:
        iteration+=1
        print('iteration: ',iteration)
        ans += num
    return ans

print(get_sum(arr))
```

- find index
- big(O) worst case scenerio
- in best case it's o(1)

```python
# linear search concept
arr=[1,2,3,4,5,6,7,15,16,17]
target=17
def find_index(arr,target):
    iteration=0
    for i in range(len(arr)):
        iteration+=1
        print('iteration',iteration)
        if arr[i]==target:            
            return i

print(find_index(arr,target))
```

# currency converters

```python
def currency_converters():
    print("---Currency converters---")
    print("1.INR to USD")
    print("2.INR to EUR")
    print("3.INR to GBP")
    print("4.USD to INR")
    print("5.EUR to INR")
    print("6.GBP to INR")
    print("7.Exit")

    INR_to_USD = 0.012
    INR_to_EUR = 0.011
    INR_to_GBP = 0.0095
    USD_to_INR = 83.0
    EUR_to_INR = 90.0
    GBP_to_INR = 105.0

    while True:
        choice=int(input('\nEnter your option (1 to 7):'))

        if choice == 7:
            print("Exiting program")
            break

        amount=float(input("Enter amount: "))

        if choice == 1:
            print("INR to USD =", amount * INR_to_USD)
        elif choice == 2:
            print("INR to EUR =", amount * INR_to_EUR)
        elif choice == 3:
            print("INR to GBP =", amount * INR_to_GBP)
        elif choice == 4:
            print("USD to INR =", amount * USD_to_INR)
        elif choice == 5:
            print("EUR to INR =", amount * EUR_to_INR)
        elif choice == 6:
            print("GBP to INR =", amount * GBP_to_INR)
        else:
            print("Invalid option. Please try again.")

currency_converters()
```
